```python
ROOT = "/content/drive/MyDrive/Upwork/"
CHUNK_DIR = ROOT + "distance_chunks/"
OUTPUT_DIR = ROOT + "distance_output/"
```

```python
import numpy as np
import pandas as pd
import os

def read_chunk_and_convert(chunk_no, format_="csv"):
  chunk_address = CHUNK_DIR + "chunk_" + str(chunk_no) + ".npy"
  chunk = np.load(chunk_address)
  chunk = pd.DataFrame(chunk)

  os.makedirs(OUTPUT_DIR, exist_ok=True)

  if format_ == "csv":
    chunk.to_csv(OUTPUT_DIR + "chunk_" + str(chunk_no) + ".csv", index=False)
  elif format_ == "xlsx":
    chunk.to_excel(OUTPUT_DIR + "chunk_" + str(chunk_no) + ".xlsx", index=False)
  else:
    print("Invalid format!")

  return "CONVERSION DONE, PLEASE CHECK YOUR OUTPUT DIRECTORY!"
```

```python
# Call the function to convert the chunk
read_chunk_and_convert(0)
```
